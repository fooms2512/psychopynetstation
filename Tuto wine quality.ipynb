```
# Tuto from https://www.youtube.com/watch?v=9snRt_5CDhs&list=PLn0OLiymPak0FvdmJx3iFL3dQywYX2kQw&index=3&ab_channel=MikeXCohen
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"

df = pd.read_csv(url, sep = ";")
df.head()
df.describe()

# print a list of unique value
print(len(pd.unique(df['pH'])))

for i in df.keys():
  print('unique value in ' + str(i) + '= ' + str(len(pd.unique(df[i]))))
```

```
#create boxplots for every columns

fig,ax = plt.subplots(1,figsize=(18,5))
ax = sns.boxplot(data=df)
ax.set_xticklabels(df.keys(),rotation=45)
plt.show()

# Remove outliers by keeping data that are bellow 200 ie we keep the non outliers ;-)
df = df[df['total sulfur dioxide']<200]
```

```
#visual exploration hue will color plot according to the wine quality
cols2plot = ['fixed acidity', 'citric acid','alcohol', 'pH', 'quality']
sns.scatterplot(x=df['residual sugar'], y=df['alcohol'])
plt.show()
sns.pairplot(df[cols2plot], kind='reg', hue='quality')
```
